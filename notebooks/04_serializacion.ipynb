```python
import joblib
import numpy as np
import pandas as pd
```

```python
# Cargar objetos generados en etapas previas
data_scaler = joblib.load('../binary/scaler_obj.joblib')
pca_model = joblib.load('../binary/pca_obj.joblib')
cluster_centers = joblib.load('../binary/cluster_centers.joblib')
print(cluster_centers.shape)
print(cluster_centers)

# Obteniendo lista de parametros
data_columns = pd.read_csv('../data/processed/data_fixed.csv').columns.to_list()
print(data_columns)
```

```python
# Revirtiendo centroides a las 16 dimensiones del perfil quimico

centroides_pca = cluster_centers.reshape((4, 10))
cluster_centers_15d = pca_model.inverse_transform(centroides_pca)
print(cluster_centers_15d)
```

```python
# Guardando objetos escenciales para uso de modelo
n_clusters = 4
n_features = 10

modelo_completo = {
    'scaler': data_scaler,
    'pca': pca_model,
    'cluster_centers_pca': cluster_centers.reshape((n_clusters, n_features)),
    'cluster_centers_15d': cluster_centers_15d,
    'columnas_originales': data_columns,
    'metadata': {
        'n_clusters': n_clusters,
        'n_features_pca': n_features, 
        'n_features_abck': 5,
        'description': 'Modelo ABC-K Perfiles Químicos Relleno Sanitario',
        'version': '0.0.1',
        'date': '2026-07-03',
        'dev': ['perez_jose', 'mikdevio'],
    }
}
```

```python
# Guardado de binario
joblib.dump(modelo_completo, '../binary/model_clasificador_lixiviados.joblib')
```
