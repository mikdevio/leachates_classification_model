```python
import pandas as pd
```

```python
data = pd.read_csv("../data/processed/data_fixed.csv")
data.head(20)
```

```python
conteo_p = data["referencia"].str.startswith('P', na=False).sum()
print(conteo_p)
conteo_l = data["referencia"].str.startswith('L', na=False).sum()
print(conteo_l)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data:
np.random.seed(1)
x = data['cadmio']
y = data['cobre']
z = data['niquel']
levels = np.linspace(z.min(), z.max(), 7)

# plot:
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x, y, 'o', markersize=2, color='black')
ax.tricontourf(x, y, z, levels=levels)

# ax.set(xlim=(-3, 3), ylim=(-3, 3))

plt.show()
```
